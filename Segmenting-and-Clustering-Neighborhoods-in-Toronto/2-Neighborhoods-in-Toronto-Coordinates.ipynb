```python
import pandas as pd
```

```python
df1=pd.read_csv('toronto_data.csv')
df1.head(10)
```

```python
df2=pd.read_csv('Geospatial_Coordinates.csv')
df2.head(10)
```

```python
#Change the column name 'Postal code' to 'PostalCode'
df2 = df2.rename(columns = {"Postal Code":"PostalCode"})
df2.columns
```

```python
#Merge two datasets using PostalCodes
df_new = pd.merge(df1, df2, on='PostalCode')
```

```python
df_new
```

```python
#Save dataframe into csv file for further use
df_new.to_csv('toronto_data_with_coordinates.csv',index=False)
```
